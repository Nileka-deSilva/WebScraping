```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('test.csv')
```

```python
df
```

```python
df["room_rate"]= df["room_rate"].str.extract('(\d+)', expand=False)
df['room_rate'] = df['room_rate'].astype(float)
```

```python
df
```

```python
res = []
 
# loop to iterate all strings
for ele in df['other_info']:
    temp = [[]]
    for char in ele:
         
        # checking for upper case character
        if char.isupper():
            temp.append([])
             
        # appending character at latest list
        temp[-1].append(char)
     
    # joining lists after adding space
    res.append(' '.join(''.join(ele) for ele in temp))
df['other_info']= res
```

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–", "•"]
for char in spec_chars:
    df['other_info'] = df['other_info'].str.replace(char, ' ')
    
```

```python
df['Breakfast Included'] = df['other_info'].str.extract(r'(Breakfast included)')
df['Availability'] = df['other_info'].str.extract(r'(Only +\d+[ \w]* left)')
df['Payment details'] = df['other_info'].str.extract(r'(No prepayment needed)')
df['Breakfast Included']= df['Breakfast Included'].replace("Breakfast included", "yes").replace(np.nan,"no")
```
